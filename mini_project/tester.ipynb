```python
# getting started with necessary imports 
import pandas as pd
import numpy as np
import pickle
import warnings 
warnings.filterwarnings('ignore')
data = pd.DataFrame([{
    'Area': 2500,
    'Build Area': 1500,
    'Bedroom': 3,
    'Bathroom': 2,
    'City': 'Kathmandu',
    'Face': 'East',
    'Address': 'Lazimpat, Kathmandu',
    'Road Type': 'Paved',
    'Floors':2,
    'Parking':2
}])
```

```python
# prediction function 
def predict_price(data, path="final_model.pkl"):
    with open(path, "rb") as f:
        pipeline = pickle.load(f)
    preds = pipeline.predict(data)
    return float(np.expm1(preds))
```

```python
pred = predict_price(data)
print("Predicted price:", pred)
```

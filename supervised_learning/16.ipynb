# Cross Validation 

- A tecnique to evaluate a model by splitting the data into a training and a test set.
- The training dataset is used to train the model, while the test set is used to evaluate the model's performace.

```python
from sklearn.datasets import load_iris 
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB 

# loading the iris dataset 
iris = load_iris() 

# initializing the Gaussian Naive Bayes classifier 
gnb = GaussianNB() 

# performing 5-fold cross-validation 
scores = cross_val_score(gnb, iris.data, iris.target, cv=5) 

# printing the cross-validation scores
print("Cross-validation scores:", scores) 
print("Mean cross-validation score:", scores.mean()) 
print("Standard deviation of cross-validation scores:", scores.std())
```

# Plotting K-fold cross validation

```python
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier 

# initializing the Decision Tree classifier 
dtc_model = DecisionTreeClassifier()

# performing k-fold cross-validation with k=7 
scores_dtc = cross_val_score(dtc_model, iris.data, iris.target, cv=7) 

# plotting the cross-validation scores 
plt.figure(figsize=(10, 6)) 
plt.plot(range(1, 8), scores_dtc, marker='o', linestyle='-', color='b') 
plt.title('K-Fold Cross-Validation Scores for Decision Tree Classifier') 
plt.xlabel('Fold Number') 
plt.ylabel('Accuracy Score')
plt.xticks(range(1, 8)) 
plt.ylim(0.8, 1)
plt.show()
```

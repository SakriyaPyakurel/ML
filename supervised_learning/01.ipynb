# Scikit learn basic Usage 

```python
# import necessary libraries 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb

# simple linear regression 
from sklearn.linear_model import LinearRegression
```

```python
# loading the data 
df = sb.load_dataset('tips') 
df.head()
```

Visualizing the feature column - total_bill and target column - tip through a scatterplot 

```python
sb.lmplot(x='total_bill',y='tip',data=df)
```

<b>Interpretations:</b><br>

i. The line going through the dataset is of the model that predicts tip(Y-axis) based on total_bill(X-axis).<br>
ii. The faint line around the model line determine the dispersion of the dataset which can create certain variations in the predictions by the different instances of the same model 

```python
# selecting the features column from the dataFrame df 
X = df[['total_bill']] 
# selecting the target column from the dataFrame df 
y = df['tip'] 
# creating the linear regression model 
model = LinearRegression() 
# fitting the model with the data 
model.fit(X, y) 
# Predicting tip based on total_bill: $ 50 
predicted_tip = model.predict([[50]]) 
print(f"Predicted tip for a total bill of $50: ${predicted_tip[0]:.2f}")
```

```python
# checking the accuracy of the model 
model_score = model.score(X, y)
print(f"Model accuracy (R^2 score): {model_score:.2f}")
```

</b>Interpretations:</b><br>

i.The accuracy of the model is too low only about 0.46 i.e 46% correct.<br>
ii. There are ways to increase this accuracy such that the model can predict more correctly and accurately.

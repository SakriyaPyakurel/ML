# Polynomial regression 

A statistical technique that models a non linear relationship between an independent and dependent variables.<br>
Performs by adding polynomial terms like x^2 and x^3 to a linear regression model

```python
# importing necessary libraries 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sb
from  sklearn.preprocessing import PolynomialFeatures 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error,mean_absolute_percentage_error
import operator 
```

```python
# creating dummy data 
np.random.seed(0) 
X = 2 - 3 * np.random.normal(0, 1, 20) 
Y = X - 2 * (X ** 2) + np.random.normal(-3, 3, 20)
# reshaping the data 
X = X[:, np.newaxis] 
Y = Y[:,np.newaxis]
sb.scatterplot(x=X.flatten(),y=Y.flatten(),color='red')
```

```python
# Transforming the data to include polynomial features with degree 1
polynomial_features= PolynomialFeatures(degree=1) 
X_poly1 = polynomial_features.fit_transform(X) 
# training linear regression model on the transformed features 
poly_model1 = LinearRegression() 
poly_model1.fit(X_poly1, Y) 
Y_poly_pred1 = poly_model1.predict(X_poly1) 
```

```python
# Plotting the results 
plt.scatter(X, Y, color='red') 
sorted_axis = operator.itemgetter(0) 
sorted_zip = sorted(zip(X,Y_poly_pred1), key=sorted_axis) 
X, Y_poly_pred1 = zip(*sorted_zip) 
plt.plot(X, Y_poly_pred1, color='blue') 
plt.title('Polynomial Regression with degree 1') 
plt.xlabel('X') 
plt.ylabel('Y') 
plt.show() 
```

```python
# Transforming the data to include polynomial features with degree 2 
polynomial_features= PolynomialFeatures(degree=2)
X_poly2 = polynomial_features.fit_transform(X)
# training linear regression model on the transformed features 
poly_model2 = LinearRegression() 
poly_model2.fit(X_poly2, Y) 
Y_poly_pred2 = poly_model2.predict(X_poly2) 
# Plotting the results 
plt.scatter(X,Y,color='red') 
sorted_axis = operator.itemgetter(0) 
sorted_zip = sorted(zip(X,Y_poly_pred2), key=sorted_axis) 
X, Y_poly_pred2 = zip(*sorted_zip) 
plt.plot(X, Y_poly_pred2, color='blue') 
plt.title('Polynomial Regression with degree 2') 
plt.xlabel('X') 
plt.ylabel('Y') 
plt.show()
```

```python
# Evaluating the models 
print('Degree 1 polynomial regression model evaluation:') 
print('Mean Squared Error:', mean_squared_error(Y, Y_poly_pred1)) 
print('R2 Score:', r2_score(Y, Y_poly_pred1)) 
print('Mean Absolute Error:', mean_absolute_error(Y, Y_poly_pred1)) 
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(Y, Y_poly_pred1)) 
print('---------------------------------------')
print('\nDegree 2 polynomial regression model evaluation:') 
print('Mean Squared Error:', mean_squared_error(Y, Y_poly_pred2)) 
print('R2 Score:', r2_score(Y, Y_poly_pred2)) 
print('Mean Absolute Error:', mean_absolute_error(Y, Y_poly_pred2))
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(Y, Y_poly_pred2))
```

# Logistic Regression - for classification

Logistic regression is used to predict a categorical value based on input features.

```python
# import necessary libraries 
import seaborn as sb 
from sklearn.linear_model import LogisticRegression 
```

```python
# import the dataset 
df = sb.load_dataset('tips') 
# creating a logistic regression model 
model = LogisticRegression() 
# selecting the features and target variable 
X = df[['total_bill', 'size']] 
y = df['smoker']
# fit the model
model.fit(X, y) 
```

```python
# making predictions with input data - total_bill: $50, size: 2
predicted_smoker = model.predict([[50, 2]]) 
print(f"Predicted smoker status for a total bill of $50 and size 2: {predicted_smoker[0]}")
```

```python
# checking the accuracy of the model 
accuracy = model.score(X, y) 
print(f"Model accuracy: {accuracy:.2f}")
```

<b>Interpretation:</b> The above logistic regression model is 66% accurate.

Logistic regression uses sigmoid function to convert the continous values into range between 0 and 1 for prediction.<br>
Lets manually implement a simple sigmoid function and calculate probabilities. 

```python
import numpy as np 
def sigmoid(x):
    """Computing the sigmoid values"""
    return 1/(1+np.exp(-x)) 

# Example of the sigmoid function 
x = np.array([0, 2, 4, 6, 8]) 
sigmoid_values = sigmoid(x)
print("Sigmoid values:", sigmoid_values)
```

<b>Interpretation:</b> The values are successfully converted into in the range between 0 and 1.

```python
# Plotting the sigmoid function 
import matplotlib.pyplot as plt 
plt.plot(x, sigmoid_values, label='Sigmoid Function') 
plt.title('Sigmoid Function Plot') 
plt.xlabel('Input Value')
plt.ylabel('Sigmoid Output') 
plt.axhline(0.5, color='red', linestyle='--', label='y=0.5')
plt.axvline(0, color='green', linestyle='--', label='x=0') 
plt.ylim(bottom=0, top=1)
plt.xlim(left=1)
plt.yticks([0,0.5,1])
plt.legend()
plt.grid()
```

<b>Interpretation:</b> According to the sigmoid line now the probabilistic values for different inputs can be predicted

```python
# Making predictions with the sigmoid function
weights = np.array([0.1, 0.2])  # example weights
bias = -0.3                     # example bias
sample = np.array([50, 2])      # total_bill=50, size=2

linear_output = np.dot(sample, weights) + bias
probability = sigmoid(linear_output)
print(f"Predicted probability of smoker: {probability:.2f}")
```

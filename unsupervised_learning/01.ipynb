# Unsupervised Machine Learning

Also known as clustering it finds patterns and relationships without any first hand knowledge of the data and it's meaning.

</b>Some commonly used unsupervised machine learning Types:</b><br>

i. K-Means clustering<br>



An unsupervised machine learning algorithm which groups unlabelled data into different clusters.

```python
# import necessary libraries 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import random
```

```python
# getting started by visualizing some data points 

x = [random.uniform(0, 100) for _ in range(10)]
y = [random.uniform(0, 100) for _ in range(10)]

plt.scatter(x, y)
plt
```

```python
# utilizing the elbow method to find the optimal number of clusters 
data = list(zip(x,y)) 
inertia = [] 

for i in range(1,len(data)):
    kmeans = KMeans(n_clusters=i, random_state=42) 
    kmeans.fit(data) 
    inertia.append(kmeans.inertia_) 

plt.plot(range(1, len(data)), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of clusters (k)') 
plt.ylabel('Inertia') 
plt
```

<b>Interpretation:</b> From the above elbow method plot the optimal k is determined to be 3.

```python
# clustering the data points using k-means with k=2
kmeans = KMeans(n_clusters=2) 
kmeans.fit(data) 
plt.scatter(x, y, c=kmeans.labels_)
```

# Pycaret 

- An open source low code machine learning library in python that streamlines machine learning algorithms.
- Acts around various ml libraries and frameworks such as scikit-learn, XGBoost, LightGBM and catboost.

```python
from pycaret.utils import version 
print("PyCaret version:", version())
```

1. Exploring data repositories

```python
from pycaret.datasets import get_data 
data = get_data('index')
```

```python
# Fetching a dataset
df = get_data('poker')
df.head()
```

2. Initialize setup

```python
from pycaret.classification import * 
clf = setup(data=df,target='CLASS',session_id=123,log_experiment=False,experiment_name='poker_class_classification')
```

3. Comparing models

```python
# comparing all models in one go
best_model = compare_models()
```

4. Create model with cross fold validation

```python
lgbm = create_model('lightgbm')
```

```python
rf = create_model('rf',fold=5)
```

5. Exploring models

```python
models()
```

```python
models(type='ensemble')
```

```python
ensemble_models = compare_models(include=models(type='ensemble').index.tolist(),fold=5)
```

6. Tune Hyperparameters

```python
tuned_lgbm = tune_model(lgbm)
```

7. Blend models

```python
blended = blend_models(estimator_list=[tuned_lgbm,rf],method='soft')
```

8. Stack Models

```python
stacked = stack_models(estimator_list=[tuned_lgbm,rf],meta_model=rf)
```

9. Analyze model

```python
plot_model(tuned_lgbm)
```

```python
plot_model(tuned_lgbm,plot='feature')
```

```python
plot_model(tuned_lgbm,plot='confusion_matrix')
```

```python
evaluate_model(tuned_lgbm)
```
